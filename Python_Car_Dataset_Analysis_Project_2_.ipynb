## 1. Instruction (Data Cleaning):








### 1.1 Based on Null Values

*   Find all null values, and replace the with mean value of that column





```python
# Find the null values
car.isnull().sum()
```

```python
#replace
car.Cylinders .fillna(car.Cylinders.mean(), inplace = True)
```
