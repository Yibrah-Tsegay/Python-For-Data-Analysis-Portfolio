**Q2) Find the number of times where the "Weather is exactly clear"?**

---



```python
##getting counts for values
Data["Weather"].value_counts()
```

```python
##Filtering
Data[Data["Weather"] == 'Clear']
```

```python
#we can also call a specific column like SQL
Data[Data.Weather == "Clear"]
```

```python
#group by
Data.groupby("Weather").get_group("Clear")
```

**Q3) Find the number od time where the wind speed was exactly 4 km/h**

---



```python
#value_counts
Data.head(2)
Data["Wind Speed_km/h"].value_counts()
```

```python
#filtering
Data.head(2)
Data[Data['Wind Speed_km/h'] == 4]
```

```python
#group by 
Data.head(2)
Data.groupby("Wind Speed_km/h").get_group(4)
```
